```python
% Robert Fortunato
% Define parameters
num_bits = 1e8;  % Number of bits to simulate
power = 1e-7;  % Received signal power
voltage = sqrt(power);  % Voltage representation of a transmitted 1
amplitude = voltage;  % BPSK mapping amplitude

% Define SNR range in dB
snr_db_values = 0:2:20;

% Store BER results
ber_bpsk = zeros(size(snr_db_values));

% Loop over different SNR values
for i = 1:length(snr_db_values)
    snr_db = snr_db_values(i);
    snr_linear = 10^(snr_db / 10);  % Convert dB to linear

    % Adjust noise variance for complex noise
    noise_variance = power / snr_linear;

    % BPSK Mapping
    bits = randi([0, 1], 1, num_bits);  % Generate random bits
    transmitted_signal_bpsk = amplitude * (1 - 2 * bits);  % BPSK mapping (0 → +amplitude, 1 → -amplitude)

    % Generate Complex Gaussian Noise
    noise_real = (1 / sqrt(2)) * sqrt(noise_variance) * randn(1, num_bits);
    noise_imag = (1 / sqrt(2)) * sqrt(noise_variance) * randn(1, num_bits);
    noise_bpsk = noise_real + 1j * noise_imag;  % Complex Gaussian noise

    % Received signal
    received_signal_bpsk = transmitted_signal_bpsk + noise_bpsk;

    % Decode bits using the **real part** of the received signal
    decoded_bits_bpsk = real(received_signal_bpsk) < 0;

    % Compute BER for BPSK mapping
    ber_bpsk(i) = sum(decoded_bits_bpsk ~= bits) / num_bits;

    % Display results
    fprintf('SNR (dB): %d, BER (BPSK): %.10f\n', snr_db, ber_bpsk(i));
end

% Plot BER vs. SNR
figure;
semilogy(snr_db_values, ber_bpsk, 'o-', 'LineWidth', 2, 'DisplayName', 'Simulated BER for BPSK');
xlabel('SNR (dB)');
ylabel('Bit Error Rate (BER)');
title(sprintf('BER vs SNR for %d bits', num_bits));
grid on;
legend;
```
